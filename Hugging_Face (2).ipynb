```python
from transformers import pipeline
```

```python
# Load the spam detection pipeline
spam_classifier = pipeline("text-classification", model="philschmid/distilbert-base-multilingual-cased-sentiment")
```

```python
txt = [
    "Congratulations! you have won a 500000 INR Amazon gift card,click here to click",
    "Hi Nihal Lets meet today at 2PM",
    "URGENT:Your gamil acct has compromised,click here to secure it "
]
results = spam_classifier(txt)
label_mapping={
    "negative":"Negative",
    "neutural":"Neutural",
    "positive":"Positive"
}

for result in results:
    label = result["label"]
    score = result["score"]
    # use the label directly from the model output as the key
    sentiment = label_mapping[label]
    print(f"Sentiment: {sentiment},Confidence Score: {score:2f}")
```

# **Text Summarization**

```python
from transformers import AutoModelForSeq2SeqLM,AutoTokenizer
```

```python
# Load a Pretrained model & tokenizer
model_name = "facebook/bart-large-cnn"
```

```python
tokenizer= AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
txt="""Generative AI is a branch of artificial intelligence that focuses on creating new content rather than just analyzing existing data.
It uses advanced machine learning models, especially deep learning and neural networks, to generate text, images, audio, video, and code.
These models learn patterns from massive datasets and use that knowledge to produce realistic and meaningful outputs.
Popular examples of generative AI include ChatGPT, DALL·E, Midjourney, and Stable Diffusion.
In natural language processing, generative AI can write essays, summarize documents, and answer questions conversationally.
In computer vision, it can create images, enhance photos, and generate realistic visual art.
Generative AI is widely used in industries such as healthcare, education, entertainment, and software development.
It helps doctors analyze medical reports, supports students with personalized learning, and assists developers in writing code.
One of the key technologies behind generative AI is the transformer architecture.
Transformers enable models to understand context and relationships within data efficiently.
Generative AI improves productivity by automating repetitive and time-consuming tasks.
However, it also raises ethical concerns related to data privacy, bias, and misinformation.
Responsible AI development is essential to ensure fair and safe usage of these systems.
Generative AI models require large computational resources and high-quality data to perform well.
Despite these challenges, continuous research is making generative AI more accurate and efficient.
It encourages creativity by helping users generate ideas, designs, and solutions."""
```

```python
inputs = tokenizer(txt, return_tensors="pt", max_length=1024, truncation=True)
# Genrate the summary
summary_ids=model.generate(inputs["input_ids"],max_length=150,num_beams=4,early_stopping=True)
summary=tokenizer.decode(summary_ids[0],skip_special_tokens=True)
print("summary:",summary)
```

Text **Translation**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
tralslation_model_name = "Helsinki-NLP/opus-mt-en-es"
translation_tokenizer = AutoTokenizer.from_pretrained(tralslation_model_name)
translation_model = AutoModelForSeq2SeqLM.from_pretrained(tralslation_model_name)

input_text="my name is Akash kumar & I like study AI"
```

```python
input_ids=translation_tokenizer.encode(input_text,return_tensors="pt")
```

```python
# genrate the translation text
translated_ids = translation_model.generate(input_ids,max_length=100,num_beams=4,early_stopping=True)
translated_text = translation_tokenizer.decode(translated_ids[0],skip_special_tokens=True)
print("Translated Text:",translated_text)
```
